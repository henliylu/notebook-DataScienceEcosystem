```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
```

```python
data  = requests.get(url).text
print(data)
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
annual_table = None
for table in soup.find_all("table", class_="historical_data_table"):
    if "Annual Revenue" in table.find("th").text:
        annual_table = table
        break
```

```python
if annual_table:
    columns = ["Year", "Revenue (Millions USD)"]
    rows = []
    for tr in annual_table.find("tbody").find_all("tr"):
        tds = tr.find_all("td")
        if len(tds) == 2:  
            year = tds[0].text.strip()
            revenue = tds[1].text.strip().replace("$", "").replace(",", "")  
            rows.append([year, revenue])
```

```python
tesla_revenue = pd.DataFrame(rows, columns=columns)
print(tesla_revenue.tail())
```

```python
import plotly.express as px
```

```python
def make_graph(data, y_col, title): 
    data.index = pd.to_datetime(data.index)
    fig = px.line(
        data, 
        x=data.index,  
        y=y_col, 
        title=title,
        labels={"index": "Date", y_col: "Price (USD)"}
    )
    fig.show()
```

```python
make_graph(history_data, y_col="Close", title="Tesla Stock Price")
```
